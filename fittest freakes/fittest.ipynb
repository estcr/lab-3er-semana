```python
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time

# Configura el navegador de Selenium
driver = webdriver.Chrome()

# URL de la página
url = "https://www.fittestfreakest.es/mujer:zapatillas"

# Abre la URL
driver.get(url)

# Espera unos segundos para que cargue el contenido
time.sleep(5)

# Obtiene el HTML de la página
html = driver.page_source

# Usa BeautifulSoup para analizar el HTML
soup = BeautifulSoup(html, 'html.parser')

# Función para extraer los datos de los productos
def extract_product_data(soup):
    df = pd.DataFrame(columns=['title', 'current_price', 'normal_price', 'link'])

    # Extraer títulos
    product_titles = soup.find_all('h2', class_='ng-star-inserted')
    # Extraer precios actuales
    current_prices = soup.find_all('span', class_='is-size-6')
    # Extraer precios normales
    normal_prices = soup.find_all('span', class_='is-size-7')
    # Extraer enlaces
    links = soup.find_all('a', class_='ng-star-inserted', href=True)

    num_items = len(product_titles)

    for i in range(num_items):
        title = product_titles[i].text.strip() if i < len(product_titles) else "N/A"
        current_price = current_prices[i].text.strip() if i < len(current_prices) else "N/A"
        normal_price = normal_prices[i].text.strip() if i < len(normal_prices) else "N/A"
        link = links[i]['href'].strip() if i < len(links) else "N/A"

        df.loc[i] = [title, current_price, normal_price, link]

    return df

# Extrae los datos y los guarda en un DataFrame
df = extract_product_data(soup)

# Cierra el navegador
driver.quit()

# Muestra el DataFrame con los datos extraídos
print(df)


```

```python
# Obtener la fecha y hora actual
fecha_hora = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")

# Crear el nombre del archivo con la fecha y hora
nombre_archivo = f"productos_fittestfreakest_{fecha_hora}.csv"

# Guardar el archivo CSV
df.to_csv(nombre_archivo, index=False, encoding='utf-8-sig')
```
